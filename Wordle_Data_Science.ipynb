```python
def count_letter_frequency(word_list):
    for word in word_list['Word List']:
        for index, letter in enumerate(word):
            letter_freq[letter][index] += 1
count_letter_frequency(word_list)
```

```python
for letter, positions in letter_freq.items():
    print(f"Letter '{letter}': {positions}")
```

```python
letter_freq_df = pd.DataFrame.from_dict(letter_freq, orient='index')
```

```python
letter_freq_df.columns = ['Position 1', 'Position 2', 'Position 3', 'Position 4', 'Position 5']
# Print the DataFrame as a table
print(letter_freq_df)
```

```python
plt.figure(figsize=(15, 10))
sns.heatmap(letter_freq_df, annot=True, cmap='YlGnBu', cbar=True)
plt.title('Letter Frequency Heatmap')
plt.xlabel('Position in Word')
plt.ylabel('Letter')
plt.show()
```

```python
#Chat GPT's work to help check if I did it right. Basically after this I will sum the values in the dictionary to check if both match.
import pandas as pd
from collections import Counter

# Load the word list from the file
with open("text_list/word_list.txt") as file:
    words = file.read().splitlines()

# Count the frequency of each letter in the word list
letter_counts = Counter("".join(words))

# Create a DataFrame to display the letter frequencies
letter_freq_df = pd.DataFrame(letter_counts.items(), columns=["Letter", "Frequency"]).sort_values(by="Frequency", ascending=False)

# Display the DataFrame
print(letter_freq_df)
```

```python
for key, value in letter_freq.items():
    total = sum(value)
    print(f"Sum of values for letter '{key}': {total}")
```
